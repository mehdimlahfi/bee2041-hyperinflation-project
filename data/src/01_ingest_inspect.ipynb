```python
import pandas as pd
from pathlib import Path

PATH = ROOT /"data"/"raw"/"inflation_worlbank.csv"

# World Bank CSV has 4 lines of metadata before the real header row.
# We need to skip them to nake the coding process easier.
df = pd.read_csv(PATH, skiprows=4, encoding="utf-8-sig")

print("Shape:", df.shape)
print("Columns (first 10):", df.columns[:10].tolist())
print("Columns (last 10):", df.columns[-10:].tolist())
print("\nHead:\n", df.head(3))
print("\nMissing values (top 10):\n", df.isna().sum().sort_values(ascending=False).head(10))
```
